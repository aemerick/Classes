```python
import numpy as np
import astropy.units as u
import matplotlib.pyplot as plt
%matplotlib inline

def leapfrog_integrate(acceleration_func,  x0, v0, dt, nsteps,args=()):
    """This function ...."""
    func = lambda x: acceleration_func(x,*args)
    
    a = np.zeros([nsteps, len(x0)])
    x = np.zeros([nsteps, len(x0)])
    v = np.zeros([nsteps, len(x0)])
    
    for i in range(nsteps):
        if i == 0:
            x[i,:] = x0[:] + v0[:]*dt
            print x[i]
            a[i,:] = func(x[i,:])
            v[i,:] = v0[:] + a[i,:]*dt
        else:    
            x[i,:] = x[i-1,:] + v[i-1,:]*dt + 0.5*a[i-1,:]*dt**2
           # print x[i]
            a[i,:] = func(x[i,:])
            v[i,:] = v[i-1,:] + 0.5*(a[i,:]+a[i-1,:])*dt
    return x,v
            

def point_mass_acceleration(x,m):
    """"Point mass at origin"""
    G = 1.
    acc = - G*m*x / np.linalg.norm(x,axis=0)**3
    return acc

def HO_acceleration(x,w):
    acc = -w**2 * x
    return acc

m = 1.
w = 2.
x,v = leapfrog_integrate(HO_acceleration,  x0 = np.array([1.,0.,0.]), v0 = np.array([0.,0.,0.]), dt = 0.2, nsteps = 1000,args=(w,))
#print orbit[0], orbit[1]

plt.figure(figsize=(5,5))
plt.plot(x[:,0], v[:,0])
plt.xlim([-2.2, 2.2])
plt.ylim([-2.2, 2.2])
plt.xlabel('x', fontsize = 20)
plt.ylabel('$p_x$', fontsize = 20)
plt.title('Integrated orbit in HO', fontsize = 16)
    
```

# Problem 2:

I choose the Miyamoto & Nagai 1975 disk as this is computationally simple and somewhat realistic:
$ \phi_{disk} = - \frac{GM_{disk}}{\sqrt{R^2 + (a + \sqrt{z^2+b^2})^2}}$

We now need to pick realistic parameters of this potential. To plot the rotation curve for this disk potential we need the acceleration.

$v_{rot} = \sqrt{acc_{disk} * R}$

$acc_{disk} =  \frac{GM_{disk}}{(R^2 + (a +b)^{2})^{1.5}} * R$

My favorite observational value of the rotation curve at the solar position is from Reid 2014 (R = 8.3 kpc and v_rot = 240)


```python
a = 6.5 #kpc
b = 0.26 #kpc
G = 4.302 * 10**-6 #kpc/Msun (km/s)^2
M = 2.395*10**11 #Msun (I made this very high)
x = np.arange(1,60,0.1) #we just do this in plane (z=0) and set y=0.
y = np.zeros(590)
z = np.zeros(590)
#r = np.sqrt(x**2+y**2+z**2)

def acc_MN_disk(x,y,z,a,b,G,M):
    """Computes acceleration due to Miyamoto Nagai 1975 disk
    Units are kpc and  M_sun"""
    acc_disk = G*M/((x**2+y**2+z**2)+(a+b)**2)**1.5 * np.sqrt(x**2+y**2+z**2)
    return acc_disk

acc_disk = acc_MN_disk(x,y,z,a,b,G,M)
r = np.sqrt(x**2+y**2+z**2)
v_rot = np.sqrt(acc_disk*r)

v_rot_sun = np.ones(len(r))*240 # to fit by eye
r_sun = np.ones(len(r))*8.3 #to fit by eye
plt.figure()
plt.plot(r,v_rot,  linewidth = 3.)
plt.plot(r,v_rot_sun, linestyle = '--', linewidth = 2.)
plt.plot(r_sun,v_rot, linestyle = '--',  linewidth = 2.)
plt.xlim([2, 60])
plt.ylim([70, 260])
plt.xlabel('R [kpc]', fontsize = 20)
plt.ylabel('v$_{rot}$ [km/s]', fontsize = 20)
```

Since there is no contribution from the bulge nor dark matter halo in this potential, the disk mass had to be unrealistically high to obtain my favorite solar velocity at its position.

```python
data = np.genfromtxt('hipparcos_rv.txt')
parallax = data[:,3]
print len(parallax)

#Now applying cuts
p = np.where(parallax > 0)

l = data[p,1]
b = data[p,2]
parallax = data[p,3]
mu_l = data[p,4]
mu_b = data[p,5]
parallax_unc = data[p,6]
mu_l_unc = data[p,7]
mu_b_unc = data[p,8]
V_mag = data[p,9]
B_V = data[p,10]
mult = data[p,11]
rv = data[p,12]
rv_err = data[p,13]

c = np.where(parallax_unc/parallax < 0.05)# & mu_l_unc/mu_l < 0.05 & mu_b_unc/mu_b < 0.05)

l = l[c]
b = b[c]
parallax = parallax[c]
mu_l = mu_l[c]
mu_b = mu_b[c]
parallax_unc = parallax_unc[c]
mu_l_unc = mu_l_unc[c]
mu_b_unc = mu_b_unc[c]
V_mag = V_mag[c]
B_V = B_V[c]
mult = mult[c]
rv = rv[c]
rv_err = rv_err[c]

d = np.where(mu_l_unc/mu_l < 0.05)

l = l[d]
b = b[d]
parallax = parallax[d]
mu_l = mu_l[d]
mu_b = mu_b[d]
parallax_unc = parallax_unc[d]
mu_l_unc = mu_l_unc[d]
mu_b_unc = mu_b_unc[d]
V_mag = V_mag[d]
B_V = B_V[d]
mult = mult[d]
rv = rv[d]
rv_err = rv_err[d]

e = np.where(mu_b_unc/mu_b < 0.05)

l = l[e]
b = b[e]
parallax = parallax[e]
mu_l = mu_l[e]
mu_b = mu_b[e]
parallax_unc = parallax_unc[e]
mu_l_unc = mu_l_unc[e]
mu_b_unc = mu_b_unc[e]
V_mag = V_mag[e]
B_V = B_V[e]
mult = mult[e]
rv = rv[e]
rv_err = rv_err[e]
# Now we are only left with 5514 stars
rv = rv[np.logical_not(np.isnan(rv))] 

f = np.where(rv == rv)
l = l[f]
b = b[f]
parallax = parallax[f]
mu_l = mu_l[f]
mu_b = mu_b[f]
parallax_unc = parallax_unc[f]
mu_l_unc = mu_l_unc[f]
mu_b_unc = mu_b_unc[f]
V_mag = V_mag[f]
B_V = B_V[f]
mult = mult[f]
rv = rv[f]
rv_err = rv_err[f]

#I also remove binary stars:
g = np.where(mult == 0)

l = l[g]
b = b[g]
parallax = parallax[g]
mu_l = mu_l[g]
mu_b = mu_b[g]
parallax_unc = parallax_unc[g]
mu_l_unc = mu_l_unc[g]
mu_b_unc = mu_b_unc[g]
V_mag = V_mag[g]
B_V = B_V[g]
mult = mult[g]
rv = rv[g]
rv_err = rv_err[g]
#I ended up with slightly less stars than you...

parallax = parallax/1000 #from mas to as
dist = 1/parallax # in pc
dist = dist/1000.# now in kpc

# Now want to convert from l,b,vr,d,mu_l,mu_b to cartesian coordinates
def helio_to_cart_GC(l,b,mul,mub,vr,d):
    """This function transforsm from Heliocentric galactic coordinates to 
    cartesian coordinates in the Galactocentric reference frame. The units
    of inputted variables should be l,b [radians], mul,mub [km/s/kpc], 
    vr [km/s], d [kpc]. """
    #LSR
    U = -11.1 #km/s
    V = -12.24 #km/s
    W = -7.25 #km/s
    rsun = -8.3 #kpc
    vcirc = 240 #km/s
    
    
    #First we want to go from spherical to cartesian
      
    x = d*np.cos(b)*np.cos(l)
    y = d*np.cos(b)*np.sin(l)
    z = d*np.sin(b)
    
    vx = x/d*vr - y*mul - z*np.cos(l)*mub
    vy = y/d*vr + x*mul - z*np.sin(l)*mub
    vz = z/d*vr + d*np.cos(b)*mub

    # Now we want to move to the GC reference frame                                        
    vy = vy + vcirc
                                         
    vx = vx + U
    vy = vy + V
    vz = vz + W
                                              
    x = x + rsun
    
    return x,y,z,vx,vy,vz

l_rad = l*(2.*np.pi/360.) # now in radians
b_rad = b*(2.*np.pi/360.)
mu_l_new = mu_l*4.74 # now in km/s/kpc
mu_b_new = mu_b*4.74 # now in km/s/kpc

x,y,z,vx,vy,vz = helio_to_cart_GC(l_rad,b_rad,mu_l_new,mu_b_new,rv,dist)
   
plt.figure(figsize = (15,8))
#plt.subplots_adjust(hspace=15)
plt.subplot(1,2,1)
plt.plot(x,y,linestyle = 'noline', marker ='.', alpha = 0.2)
plt.xlim([-8.4,-8.2])
plt.ylim([-0.1,0.1])
plt.xlabel('$x$ [kpc]', fontsize = 20)
plt.ylabel('$y$ [kpc]', fontsize = 20)
plt.subplot(1,2,2)
plt.plot(x,z,linestyle = 'noline', marker ='.', alpha = 0.2)
plt.xlim([-8.4,-8.2])
plt.ylim([-0.1,0.1])
plt.xlabel('$x$ [kpc]', fontsize = 20)
plt.ylabel('$z$ [kpc]', fontsize = 20)


plt.figure(figsize = (15,6))
#plt.subplots_adjust(hspace=15)
plt.subplot(1,2,1)
plt.plot(vx,vy,linestyle = 'noline', marker ='.', alpha = 0.2)
plt.xlim([-150,150])
plt.ylim([80,360])
plt.xlabel('$v_x$ [km/s]', fontsize = 20)
plt.ylabel('$v_y$ [km/s]', fontsize = 20)
plt.subplot(1,2,2)
plt.plot(vx,vz,linestyle = 'noline', marker ='.', alpha = 0.2)
plt.xlim([-150,150])
plt.ylim([-150,150])
plt.xlabel('$v_x$ [km/s]', fontsize = 20)
plt.ylabel('$v_z$ [km/s]', fontsize = 20)
```
