```python
def leapfrog_integrate_3d(acceleration_func, x0, y0, z0, vx0, vy0, vz0, 
                         dt, nsteps, args=()):
    """
       Accepts a general acceleration function with additional arguments supplied
       by args. Integrates this function in the leapfrog scheme given initial conditions
       on x0 and v0, the specified timestep dt, and the number of steps to integrate. An array
       of initial conditions can be passed to integrate multiple orbits.
       
    Parameters
    ---
       acceleration_func : function
           acceleration function with additional arguments args()
       x0 : float, array
           float or array of initial positions/ Must be same size as v0
       v0 : float, array
           float or array of initial velocities. Must be same size as x0
       dt : float
           fixed time step size
       nsteps : int
           number of time steps to go through
       accel : optional, bool
           If true, returns the values of acceleration at each sampled x. Default is False
    Returns
    ---
       x : array
           Solved position of orbit
       v  :array
           solved velocity of orbit
       a : array 
           If accel is true, returns acceleration at each point of orbit
    """
    
    # get the acceleration function
    a = lambda t, x, y, z, coord: acceleration_func(t, x, y, z, coord, *args)
    
    # make sure 
    ninit = np.size(x0)
    if ninit != np.size(vx0) or ninit != np.size(y0) or ninit != np.size(z0)\
         or ninit != np.size(vy0) or ninit != np.size(vz0):
        print "ERROR: must have same number of initial conditions on all position and velocity vectors"
        return 
    
    # initialize the final arrays
    x, vx = np.zeros((ninit,nsteps)), np.zeros((ninit,nsteps))
    y, vy = np.zeros((ninit,nsteps)), np.zeros((ninit,nsteps))
    z, vz = np.zeros((ninit,nsteps)), np.zeros((ninit,nsteps))
                             
    # set the initial conditions for position and velocity
    x [:,0] =  x0
    vx[:,0] = vx0
    y [:,0] =  y0
    vy[:,0] = vy0
    z [:,0] =  z0
    vz[:,0] = vz0

    # now integrate! This is the synchronized method where x and v step forward together in
    # time, rather than v moving in half-steps
    # this is an array style operation where it steps through all initial conditions together.
    
    t = np.zeros(nsteps)

    for i in np.arange(nsteps - 1) + 1:
        t[i] = t[i-1] + dt

        # update position vectors using full step timesteps 
        x[:,i]  =  x[:,i-1] + vx[:,i-1] * dt +\
                          0.5 * a(t[i-1], x[:,i-1], y[:,i-1], z[:,i-1], 'x') * dt**2
        y[:,i]  =  y[:,i-1] + vy[:,i-1] * dt +\
                          0.5 * a(t[i-1], x[:,i-1], y[:,i-1], z[:,i-1], 'y') * dt**2
            
        z[:,i]  =  z[:,i-1] + vz[:,i-1] * dt +\
                          0.5 * a(t[i-1], x[:,i-1], y[:,i-1], z[:,i-1], 'z') * dt**2
            
            
            
        # update velocity vecotrs using full step time steps    
        vx[:,i] = vx[:,i-1] +\
                          0.5 * ( a(t[i-1], x[:,i-1], y[:,i-1], z[:,i-1], 'x') +\
                                a(t[i],   x[:,i],   y[:,i],   z[:,i], 'x') ) * dt
           
        vy[:,i] = vy[:,i-1] +\
                          0.5*( a(t[i-1], x[:,i-1], y[:,i-1], z[:,i-1], 'y') +\
                                a(t[i],   x[:,i],   y[:,i],   z[:,i], 'y') ) * dt

        vz[:,i] = vz[:,i-1] +\
                          0.5*( a(t[i-1], x[:,i-1], y[:,i-1], z[:,i-1], 'z') +\
                                a(t[i],   x[:,i],   y[:,i],   z[:,i], 'z') ) * dt
        

    return t, x, y, z, vx, vy, vz
```

```python
def potential(potential_type):
    """
    Returns the desired potential function, determined by potential type as
    the isochrone, bar, or the sum of the two
    """
    
    
    def _isochrone(t, x, y, z, b = 1., GM = 1):
        return -GM / ( b + (b*b + (x**2 + y**2 + z**2))**0.5)
    
    def _bar(t, x, y, z, a = 2.6, epsilon = 0.5, omega = 0.2):
        
        phi = np.arctan2(y , x)
        r   = (x**2+y**2+z**2)**0.5
        ra  = r / a
        
        ra_1 = ra[ra<1]
        ra_2 = ra[ra>=1]
        psi = np.zeros(np.size(x))
        psi[ ra < 1 ] = -0.53 * (ra_1)**2 * (1.0 - 2.5*ra_1**2 + 2.1875*ra_1**4 - 0.65625*ra_1**6)
        psi[ ra >= 1] = 0.017 * ra_2** (-4.0)
        
        psi *= epsilon

        return psi * np.cos(2.0*(phi - omega*t))
    
    
    def _total_potential(t, x, y, z, b = 1., GM=1, a=2.6, epsilon=0.5, omega=0.2):
        
        return _isochrone(t,x,y,z,b,GM) + _bar(t,x,y,z,a,epsilon,omega)
    
    
    
    pot_dict = {'isochrone': _isochrone, 'bar': _bar, 'total': _total_potential}
    
    return pot_dict[potential_type]
    
```

```python
class Satellite:   
    """
    Satellite object for doing the orbiting satellite galaxy perturbations. Initialized
    with initial positions and velocities of the satellite, along with a GM and ID
    """
    
    def __init__(self, x0, y0, z0, vx0, vy0, vz0, GM, ID=None):
        """
        Initialize with positions, velocities, mass, ID
        Also calcuates the initial inclination angle
        """
        self.x0 = x0
        self.y0 = y0
        self.z0 = z0
        self.vx0 = vx0
        self.vy0 = vy0
        self.vz0 = vz0
        self.GM  = GM
        
        self.ID = ID
        
        
        self.i = np.arccos((self.vx0**2+self.vy0**2)**0.5/(self.vx0**2+self.vy0**2+self.vz0**2)**0.5)
        
    def integrate(self, accel_func, dt, nsteps):
        """
        Integrate to find the orbit of the satellite. This is necessary before using
        the satellite in the acceleration function for integrating the disk stars. Stores
        the satellite positions and velocities, time, dt, and number of timesteps
        """
        
        t,x,y,z,vx,vy,vz = leapfrog_integrate_3d(accel_func, self.x0, self.y0, self.z0,
                                               self.vx0, self.vy0, self.vz0, dt, nsteps)
        self.t  = t
        self.dt = dt
        self.nsteps = nsteps
        
        self.x = x[0]
        self.y = y[0]
        self.z = z[0]
        self.vx = vx[0]
        self.vy = vy[0]
        self.vz = vz[0]
        
        
        self.apocenter = np.max(self.r())
        self.pericenter = np.min(self.r())
        
    def r(self, t=None, n=None):
        
        if t == None and n == None:
            return (self.x**2+self.y**2+self.z**2)**0.5 # return all r values
        elif n == None:
            n = t / self.dt
        
        return (self.x[n]**2 + self.y[n]**2 + self.z[n]**2)**0.5
        
            
```

```python
def acceleration(potential_type):
    """
    Returns the desired potential function, determined by potential type as
    the isochrone, bar, or the sum of the two
    """
    
    
    def _isochrone(t, x, y, z, coord, b = 1., GM = 1.):
        
        r = (x**2 + y**2 + z**2)**0.5
        
        if coord == 'x':
            val = x
        elif coord == 'y':
            val = y
        elif coord == 'z':
            val = z
            
        return  -1.0* GM * val / ( (b**2+r**2)**0.5 * ((b**2+r**2)**0.5 + b)**2)     
            
            
       # return 1.0 * val * GM / ( (b**2 + r**2)**0.5 * ((b**2+r**2)**0.5 + b**2)**2)
        
            
    def _bar(t, x, y, z, coord, a = 2.6, epsilon = 0.5, omega = 0.2, ramp=[False,0]):
        
        phi = np.arctan2(y , x)
        
        r   = (x**2+y**2+z**2)**0.5
        ra  = r/a
        
        if coord == 'x':
            val = x
            dcos = 2.0 * y * np.sin(2.0*(phi - omega*t))/(x**2 + y**2)
        elif coord == 'y':
            val = y
            dcos = 2.0 * x * np.sin(2.0*(-phi + omega*t))/(x**2 + y**2)

        else:
            val = z
            dcos = np.zeros(np.size(x))
        
        psi = np.zeros(np.size(x))
        dpsi = np.zeros(np.size(x))
        
        cut1 = ra<1
        cut2 = ra>=1
        
        ra_1 = ra[cut1]
        ra_2 = ra[cut2]
        psi[cut1] = -0.530 * (ra_1)**2 * (1.0 - 2.5*ra_1**2 + 2.1875*ra_1**4 - 0.65625*ra_1**6)
        psi[cut2] =  0.017 * (ra_2)**(-4.0)
        
        
        dpsi[cut1] = 4.0 * val[cut1] * psi[cut1]/(ra_1*a) +\
                        -0.53*ra_1**2 *val[cut1] * (-10.0*ra_1/a + 17.5*ra_1**3/a - 7.875*ra_1**5/a)
        dpsi[cut2] = -0.136 * val[cut2] * ra_2**(-5.0) / a
        
    
        if ramp[0] and t / ramp[1] <= 1000:
            n       = t / ramp[1]
            epsilon = epsilon * (n / 1000.0)
       
        psi  *= epsilon
        dpsi *= epsilon

        return -1.0*(psi * dcos + dpsi * np.cos(2.0*(phi - omega*t)))
    
    def _pert(t, x, y, z, coord, satellite=None):
        """
        perturnbing acceleration
        """
        
        n = t / satellite.dt
        
        d = ((x - satellite.x[n])**2 + (y-satellite.y[n])**2 + (z-satellite.z[n])**2)**0.5
        
        if coord == 'x':
            val = x - satellite.x[n]
        elif coord =='y':
            val = y - satellite.y[n]
        elif coord =='z':
            val = z - satellite.z[n]
            
        return -1.0 * satellite.GM * val / d**3
        
        
    
    def _total_accel(t, x, y, z, coord, b = 1., GM=1., a=2.6, epsilon=0.5, omega=0.2, ramp=[False,0]):
        
        return _isochrone(t,x,y,z,coord, b,GM) + _bar(t,x,y,z,coord, a,epsilon,omega,ramp)
    
    def _isochrone_pert(t, x, y, z, coord, b=1.,GM=1., satellite=None):
        return _isochrone(t,x,y,z,coord,b,GM) + _pert(t,x,y,z,coord,satellite)
    
    pot_dict = {'isochrone': _isochrone, 'bar': _bar, 'total': _total_accel, 
                'iso_sat': _isochrone_pert}
    
    return pot_dict[potential_type]

def vc_isochrone(r, b = 1.0, GM = 1.0):
    """
    Velocity curve for the isochrone potential
    """
    
    vc = r * r * GM / ( (b**2 + r**2)**0.5 * ((b**2+r**2)**0.5 + b)**2)
    
    return vc**0.5

def vc_isochrone_dr(r, b=1.0,GM=1.0):
    """
    Returns the first derivative (in r) of v_c^2 for the isochrone potential
    """
    
    return vc_isochrone(r,b,GM)**2 * (2.0/r - r/(b**2+r**2) - \
                              2.0*r/(b**2 + r**2 + b*(b**2+r**2)**0.5))
    
   


```

```python
def star_pos(nstars, ndim, rmax, b=1.,GM=1.):
    
    
    z = np.zeros(nstars)
    vz = np.zeros(nstars)
    
    r   = np.random.rand(nstars)*rmax**ndim # uniform density... ndim
    r   = r**(1.0/(ndim*1.0))
    phi = np.random.rand(nstars)*2.0*np.pi
    z = np.zeros(nstars)

    x = r*np.cos(phi)
    y = r*np.sin(phi)
    
    vc     = vc_isochrone(r, b=b, GM=GM)
    
    print "vc", np.min(vc), np.max(vc)
    
    vx = -vc*np.sin(phi)
    vy =  vc*np.cos(phi)
    
    
    return x,y,z,vx,vy,vz
```

```python
def co_rotation(vcfunc, omega_bar, r0=2., args=()):
    """
    Calculates the co rotation radius given vc and omega
    """
    
    # function to find the root for is circular velocity / r - the pattern speed
    f = lambda r: ((vcfunc(r, *args)) / r) - omega_bar 

    cr = opt.newton(f, r0)
    
    return cr
```

```python
def lindblad_resonances(vcfunc, vcsqr_dr, omega_bar, 
                        rmax, m=2., nsamples=10, unique_tol = 1.0E-5, plot_freq=False, args=()):
    """
    Returns any lindblad resonances given the circular velocity curve function and the 
    derivative of the circular velocity squared, in addition to the bar pattern speed 
    over the range of rmin and rmax. The function samples nsamples radii over the range [rmin,rmax]
    and returns any non-zero, unique roots found
    
    
    """

    # kappa frequencies... last term is omega_o**2  .. first is 1/r * d(v_c^2)/dr
    kappa = lambda r: vcsqr_dr(r,*args) / r - (vcfunc(r,*args)/r)**2
    
    fouter     = lambda r: kappa(r)/(m*1.) - (vcfunc(r,*args)/r - omega_bar)
    finner     = lambda r: -kappa(r)/(m*1.) - (vcfunc(r,*args)/r - omega_bar)

    if plot_freq:
        plt.figure(figsize=(6,6))
        r = np.linspace(0.01,rmax*2.,100)
        om   = vcfunc(r,*args)/r
        omup = om + kappa(r)/(m*1.)
        omdn = om - kappa(r)/(m*1.)
        plt.title('m = %2i'%(int(m)))
        plt.plot(r,omup,label=r'$\Omega$ + $\kappa$/m', color='red',lw=1.5)
        plt.plot(r,omdn,label=r'$\Omega$ - $\kappa$/m', color='blue',lw=1.5)
        plt.plot(r,omega_bar*np.ones(np.size(r)),label=r'$\Omega_{bar}$', color='black',lw=1.5)
        plt.xlabel('r (kpc)')
        plt.ylabel('Frequency')
        plt.legend(loc='best',fancybox=True)
        plt.savefig('plots/frequencies.png')
        #plt.close()
    
    
    fdict = {'finner':finner,'fouter':fouter}
    
    final_radii = {}
    
    for fname in fdict:
        sradii = np.linspace(0.01, rmax, nsamples)
        roots  = np.ones(nsamples)*-1.0

        for i in np.arange(nsamples):
            try:
                roots[i] = opt.newton(fdict[fname], sradii[i])
            except:
                print "No root found for " + fname 


        roots = roots[(roots>=0)*(roots<rmax*10)]
        
        if np.size(roots) > 1:
            final_roots = [roots[0]]

            for root in roots:

                unique = True
                for fr in final_roots:
                    if np.abs((root - fr)/fr) < unique_tol:
                        unique = False

                if unique:
                    final_roots.append(root)             
        else:
            final_roots = []
        
        final_radii[fname] = np.array(final_roots)
            
    return final_radii['finner'], final_radii['fouter']    
    
```

```python
def find_period(pos, dt, t = None, axis=0):
    """
    takes series of positions and time in between each position to find the period
    
    dt if constant time step sizes between position points
    t array of sime size as pos giving the time at each position for an arbirary dt.
    
    """
           
    index_maxima = signal.argrelmax(pos, axis=axis)
    
    if np.size(t) == np.size(pos):
        time    = t[index_maxima]
        periods = t[1:] - t[:-1]
    else:    
    
        period = np.zeros(np.size(pos,axis=0))
        for i in index_maxima[0]:
            period[i] = np.average( (index_maxima[1][index_maxima[0]==i][1: ] -\
                                     index_maxima[1][index_maxima[0]==i][:-1] ) * dt)
    
    
    return period
```

## Interpreting the results...

The final disk morphology depends strongly on the inclination angle of the satellite orbit to the disk of the galaxy. In each set of runs, from c1 - c5 and e1 - e5, the inclination angle ranges from $\pi$/2 to 0. In the 'cX' tests, a lower inclination angle causes greater disturbance to the disk, while the effects are less pronounced at i = $\pi$/2. Although all of the 'cX' orbits have the same perihelion and aphelion (about 15 and 17 respectively), as i $\rightarrow$ 0, the satellite spends more time closer to more stars (those towards the edges of the disk in particular). The inclination angle also changes the shape of the final galaxy. This is seen in the xy projection, but is confirmed in the 3D plot. The final galaxy is elongated in the plane of the satellite's orbit. For the 'c1' case epsecially, the 3D projection shows a central disk that remains relatively undisturbed by the satellite surrounded by a ring of points that is inclined towards the satellite relative to the disk.

For the elliptical orbits again, there are less stars remaining in the disk as i $\rightarrow$ 0, with the i = 0 case devoid of any recognizeable disk. The elliptical orbits in general are dramatically more disturbing than the circular as the satellite spends more time close to stars i n the disk, and plunges deep into the potential of the galaxy at a pericenter of about 3.4. As in the circular orbits case, the morphology of the final disk(ish) structure rotates with the inclination angle of the satellite. 

The less massive runs (obviously) produce smaller disturbances in the disk. In the 'c1' case, the less massive satellite is barely noticeable, with only some percent-level changes to the outer edge of the disk. Changes become more noticeable, again, as i $\rightarrow$ 0, with the perturbed stars rotating with the inclination angle of the satellite. As i $\rightarrow$ 0, the disk contains noticeable warps and an elongated puffy region of stars around the disk. When i = 0, this puffiness disappears (as it is the result of a nonzero i). The location of the satellite is noticeable as it appears to be trailed by stars "leaking" out of the edge of the galaxy's disk, some of which are flung out and unbound.

In the ellipictal orbits, the less massive satellite keeps a noticeable disk in the galaxy at all orbit inclinations. Compared to the 'cX' runs, the 'eX' runs produce a puffier, prolate spheroid looking thing (at least in projection).

These computations are very unrealistic for several reasons. The three largest are that we are modelling the satellite as a moving point mass, the galaxy potential is an isochrone, and not a disk, and we are treating the points as collisonless. Since the "points" represented stars, having them be collisonless isn't really all that bad. However, doing a "real" N-body simulation of the effects of a satellite orbiting a galaxy would require massive particles and some level of self-gravity for these massive (presumably dark matter) particles. The potential of both the galaxy and its satellite would have to be computed "live" from the particle distribution to model things properly. This is because a real merger would produce a time varying potential in both the galaxy and the satellite. In many cases, tidal effects in a non-point mass satellite will, over time, weaken its potential. This effect will be less dramatic in the more massive host galaxy, but still relevant.
